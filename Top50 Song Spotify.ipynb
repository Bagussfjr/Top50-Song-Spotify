https://www.kaggle.com/leonardopena/top50spotify2019

1. ID
2. Track.Name => Name of the Track
3. Artist.Name => Name of the Artist
4. Genre => The genre of the track
5. Beats.Per.Minute => The tempo of the song.
6. Energy => The energy of a song - the higher the value, the more energtic. song
7. Danceability => The higher the value, the easier it is to dance to this song.
8. Loudness..dB.. => The higher the value, the louder the song.
9. Liveness => The higher the value, the more likely the song is a live recording.
10. Valence. => The higher the value, the more positive mood for the song.
11. Length. => The duration of the song.
12. Acousticness.. => The higher the value the more acoustic the song is.
13. Speechiness. => The higher the value the more spoken word the song contains.
14. Popularity => The higher the value the more popular the song is.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
kolom = ['Nomor', 'Track Name', 'Artist Name', 'Genre', 'Beats per Minute',
        'Energy', 'Danceability', 'Loudness', 'Liveness', 'Valence', 'Lenght',
        'Acousticness', 'Speechiness', 'Popularity']
data = pd.read_csv('top50.csv', sep = ',', names = kolom, skiprows = 1, encoding = 'cp1252')
data.head(10)
```

```python
data.shape
```

```python
data.describe()
```

```python
y = pd.DataFrame(data.Genre.value_counts()).reset_index().head()
y['Total'] = y['Genre']
y['Genre'] = y['index']
data_spotify = y[['Genre', 'Total']]
data_spotify.head()
```

```python
bar_data = sns.barplot(x = "Genre", y = "Total", data = data_spotify)
bar_data.set_xticklabels(bar_data.get_xticklabels(), rotation = 90) # Mengubah tulisan x menjadi bentuk vertical
print(bar_data)
```
